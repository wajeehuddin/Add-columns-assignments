```python
import pandas as pd
```

```python
data = pd.read_csv("Pokemon.csv")
data.head(n=10)
```

# Hp column:

```python
HP_mean = data["HP"].mean()
def set_HP(val):
    if val < HP_mean:
        return "HP Low"
    elif val == HP_mean:
        return "HP neutral"
    else:
        return "HP High"
```

```python
data["HP_high_low"] = data["HP"].apply(set_HP)
data
```

```python
Speed_mean = data["Speed"].mean()
def set_Speed(val):
    if val < Speed_mean:
        return "Speed Low"
    elif val == Speed_mean:
        return "Speed neutral"
    else:
        return "Speed High"
```

```python
data["Speed_High_Low"] = data["Speed"].apply(set_Speed)
data
```
